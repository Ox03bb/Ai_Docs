## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pandas as pd
```

## 2- import data

```python
path = './data/Housing_3par.csv'

data = pd.read_csv(path)

data.head()
```

```python
X = data['area'].values
Y = data['price'].values
Z = data['bedrooms'].values

X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
```

```python
#3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z, c='r', marker='o')
ax.set_xlabel('Area')
ax.set_ylabel('Price')
ax.set_zlabel('Bedrooms')
```

```python
nx = X*1.0/np.max(X)
ny = Y*1.0/np.max(Y)
nz = Z*1.0/np.max(Z)
```

```python
#3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(nx, ny, nz, c='r', marker='o')
ax.set_xlabel('Area')
ax.set_ylabel('Price')
ax.set_zlabel('Bedrooms')
```
