# TenserFlow liniar regression

 ## Package 

```python
import numpy as np

import matplotlib.pyplot as plt

import pandas as pd

import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
from tensorflow.keras.activations import sigmoid
```

## Data

```python
data = pd.read_csv('./data/Housing_2par.csv')

data.head()
```

```python
data.plot.scatter(x='area', y='price',s=10, c='black')
```

## Normalize data

```python
x_train = data['area'].values
y_train = data['price'].values

std_X = np.std(x_train)
mean_X = np.mean(x_train)

nrm_x = (x_train - np.mean(x_train)) / np.std(x_train)
nrm_y = (y_train - np.mean(y_train)) / np.std(y_train) 

plt.scatter(nrm_x, nrm_y, s=10, c='black')
```

## nueural network

```python
model = Sequential() 
model.Dense(units=1, activation='linear')
```

```python
model.get_weights()
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
w, b= model.get_weights()
print(f"w = {w}, b={b}")
```

```python
plt.scatter(nrm_x, nrm_y, s=10, c='black')
plt.plot(nrm_x, model(nrm_x), c='red')
```
