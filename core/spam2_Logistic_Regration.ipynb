# Spam Classification

# I- Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import confusion_matrix

import seaborn as sns

%matplotlib inline
```

## II- Data

```python
data = pd.read_csv('./data/emails.csv')

data.head()
```

```python
x_data = data.drop('Prediction', axis=1)
x_data = x_data.drop('Email No.', axis=1).values
y_data = data['Prediction'].values
```

## III- Data Preprocessing

```python
model = LR()
model.fit(x_data, y_data)
```

```python
print(model.score(x_data, y_data))
```

```python
y_pred = model.predict(x_data)
```

```python
acc = confusion_matrix(y_data, y_pred)

sns.heatmap(acc,annot=True,fmt="d")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.text(0.75, 2.5, f'acuracy: {model.score(x_data, y_data)*100:.2f}%', fontsize=12)
plt.show()
```
