# Logistic Regression

## I-Package

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LogisticRegression as LR

%matplotlib widget
```

## II- Data

```python
data = pd.read_csv('./data/spam_email_dataset.csv')

data.head()
```

```python
x_data = data.values[:,6:15]
y_train= data.values[:,-1:]

x_train = sk.preprocessing.scale(x_data)
y_train = y_train.flatten()

```

## III- visualisation of the data

```python
labels = ['word_freq_make', 'word_freq_address', 'word_freq_all', 'word_freq_3d', 'word_freq_our', 'word_freq_over', 'word_freq_remove', 'word_freq_internet', 'word_freq_order']
# colors = ['r', 'g', 'b', 'y', 'c', 'm', 'k', 'r', 'g']

plt.figure(figsize=(10, 7))

for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.scatter(x_train[:,i], y_train,alpha=0.005 , label=labels[i])
    plt.legend()
```

## IV- Model

```python
# Convert the target variable to discrete classes
label_encoder = sk.preprocessing.LabelEncoder()

# y_train = label_encoder.fit_transform(y_train)

# Train the model
model = LR()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```
